### Уровень 1:
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
def f_(x):
    return (np.sin(2*x+1))**5
```

```python
x0 = np.arange(-5, 6, 0.01)
x = [derivative(f_, x0[i], dx=1e-6) for i in range(len(x0))]
x
```

```python
x2 = [derivative(f_, x0[i], n=2, dx=1e-6) for i in range(len(x0))]
x2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x0, x, '-', color='b', label = 'Первая производная')
```

```python
plt.plot(x0, x2, '-', color='b', label = 'Вторая производная')
```
