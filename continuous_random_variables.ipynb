```python
from scipy.stats import expon
from scipy.stats import norm
from scipy.stats import t
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
```

Нормальное распределение

```python
mu=2.5
sigma = 0.25
# зададим нормальнораспределенную случайную величину
norm_rv = sts.norm(loc=mu, scale=sigma)
# сгенирируем 10 значений 
norm_rv.rvs(size=10)
```

```python
norm_rv_2 = sts.norm(loc=2, scale=0.5)
norm_rv_3 = sts.norm(loc=2.5, scale=0.5)
norm_rv_4 = sts.norm(loc=4, scale=0.5)
norm_rv_5 = sts.norm(loc=7, scale=1)
```

```python
x= np.linspace(0,5,100)
x5= np.linspace(0,10,100)
pdf = norm_rv.pdf(x)
pdf2 = norm_rv_2.pdf(x)
pdf3 = norm_rv_3.pdf(x)
pdf4 = norm_rv_4.pdf(x)
pdf5 = norm_rv_5.pdf(x5)
plt.plot(x, pdf)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
plt.plot(x, pdf4)
plt.plot(x5, pdf5)
```

Экспоненциальное распределение

```python
lam=0.1
expon_rv = sts.expon(lam)
expon_rv_2 = sts.expon(1)
expon_rv_3 = sts.expon(-0.1)
expon_rv_4 = sts.expon(5)
expon_rv_5 = sts.expon(2.5)
```

```python
x = np.linspace(-0.5, 8, 1000)
pdf = expon_rv.pdf(x)
pdf_2 = expon_rv_2.pdf(x)
pdf_3 = expon_rv_3.pdf(x)
pdf_4 = expon_rv_4.pdf(x)
pdf_5 = expon_rv_5.pdf(x)
plt.plot(x, pdf)
plt.plot(x, pdf_2)
plt.plot(x, pdf_3)
plt.plot(x, pdf_4)
plt.plot(x, pdf_5)
```

- Задание 3

Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
red = 54
yellow = 4
green = 28
```

```python
# формула шенона
import math
p_red = red/(red+yellow+green)
p_yellow = yellow/(red+yellow+green)
p_green = green/(red+yellow+green)
I = -(p_red*math.log2(p_red))+(-(p_yellow*math.log2(p_yellow)))+(-(p_green*math.log2(p_green)))
print(f'Сообщение несет {I} бит')
```
